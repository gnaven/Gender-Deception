## Correlations

```python
# Use itertools.combinations to get all unique correlations from pandas 
# own correlation matrix .corr(), generate list of lists and feed it back 
# into a DataFrame in order to use '.sort_values'. Set ascending = True to 
# display lowest correlations on top

# corrank takes a DataFrame as argument because it requires .corr().
from itertools import combinations
from itertools import product
from itertools import permutations

def corr_clean(df):
    """ returns a df of all nonredundant pairwise correlations. 
    only non-nan producing pair correlations returned. nans are individually 
    excluded. 
    """
    from itertools import combinations
    df_corr = df.corr() # prevent recomputation
    r_pairs = [[(i,j),df_corr.loc[i,j]] for i,j in list(combinations(df_corr, 2))]
    df_r_clean = pd.DataFrame(r_pairs, columns=['pairs','corr'])    
    df_r_clean['abs'] = df_r_clean['corr'].abs()    
    df_r_clean.sort_values(by='abs',ascending=False, inplace=True)
    return df_r_clean

def print_correlations(df):
    """ displays a set of correlations between field groups """

    field_pairs = {'(AUS,LIWC)':(AUS,LIWC)}

    df_corr = df.corr()
    permutes = list(permutations(df_corr, 2))
    r_pairs = [[(i,j),df_corr.loc[i,j]] for i,j in permutes]
    df_r_clean = pd.DataFrame(r_pairs, columns=['pairs','corr']) 
    df_r_clean['abs'] = df_r_clean['corr'].abs()    
    df_r_clean.sort_values(by='abs',ascending=False, inplace=True)
    
    for field_pair_str, field_pair in field_pairs.items():
        pairs = product(field_pair[0],field_pair[1])
        print("\n" + field_pair_str)
        display(df_r_clean[df_r_clean['pairs'].isin(pairs)][0:10])
        
print_correlations(df)
print_correlations(df[df['truth_val'] == 1])
print_correlations(df[df['truth_val'] == 0])
```
