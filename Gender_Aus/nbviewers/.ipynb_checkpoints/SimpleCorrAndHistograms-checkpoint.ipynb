# Basic stats of AU data

This is a simple analysis of all the face frame dta that goes into clustering. It includes both witness, interrogator, during both truth, bluff, relevant and irrelevant questioning. 

* %'s of zero frames
* correlations
  * AU_c 
  * AU_r
* histograms
  * all data
  * zeros removed

In summary, the data is 1.2M frames. The highest AU_r correlation is between (AU06_r x	AU12_r) (0.720) and (AU01_r x AU02_r) (0.702). The distributions are very heavy at zero, as well as close but not equal to zero. Some AUs have an additional mode around [0.5-1] (AU10,12,14 and to a lesser degree AU6, 7, 17). The shapes of the histograms form the basis of a clustering algorithm with a better basis distribution.

## Correlations

```python
def get_redundant_pairs(df):
    '''Get diagonal and lower triangular pairs of correlation matrix'''
    pairs_to_drop = set()
    cols = df.columns
    for i in range(0, df.shape[1]):
        for j in range(0, i+1):
            pairs_to_drop.add((cols[i], cols[j]))
    return pairs_to_drop

def get_top_abs_correlations(df, n=5):
    au_corr = df.corr().unstack()
    labels_to_drop = get_redundant_pairs(df)
    au_corr = au_corr.drop(labels=labels_to_drop).reset_index()
    au_corr.columns = ['0','1','r']
    au_corr['abs(r)'] = au_corr['r'].abs()
    au_corr.sort_values(by='abs(r)',ascending=False, inplace=True)
    return au_corr[0:n]
#============================================================
print("Top AU_c Correlations")
display(get_top_abs_correlations(df_c, 15))
print()
print("Top AU_r Correlations")
display(get_top_abs_correlations(df_r, 15))
```

## Histograms

```python
# 1D histograms
for feature in df_r.columns:
    plt.hist(df_r.loc[:,feature], 200, color='green', \
         histtype='bar', ec='black', normed=1)
    plt.xlabel(feature)
    plt.ylabel('frequency')
    plt.show()
```

### 1D Histograms with zero removed

These plots show tbhe interesting features of the distributions since the zero spike is not masking anymore. Its important to note that there is still significant concentration around zero. Also, there are multiple modes with AUs 10,12,14

```python
# AU histograms when ignoring zeros
for feature in df_r.columns:
    df_x = df_r[df_r[feature] != 0][feature]
    plt.hist(df_x, 200, color='green', \
         histtype='bar', ec='black', normed=1)
    plt.xlabel(feature)
    plt.ylabel('frequency')
    plt.show()
```

### 2D histograms for smile-related AUs

```python
# pairwise 2D histograms for smile-related AUs
from matplotlib.colors import LogNorm
from itertools import combinations
good_features = ['AU06_r','AU07_r','AU10_r','AU12_r','AU14_r']

feature_combo_list = list(combinations(good_features,2))        
for feature_subset in feature_combo_list:
    plt.figure(figsize=(4,4))
    h = plt.hist2d(df.loc[:,feature_subset[0]], df.loc[:,feature_subset[1]], bins=40, norm=LogNorm())
    #plt.hist2d(df.loc[:,feature_subset[0]], df.loc[:,feature_subset[1]], bins=40)
    plt.xlabel(feature_subset[0])
    plt.ylabel(feature_subset[1])
    #plt.colorbar()
    plt.show()
```

```python
# 2D hist with no origin zero
from matplotlib.colors import LogNorm
from itertools import combinations

good_features = ['AU06_r','AU07_r','AU10_r','AU12_r','AU14_r']

feature_combo_list = list(combinations(good_features,2))        
for feature_subset in feature_combo_list:
    plt.figure(figsize=(4,4))
    #df_x = df[(df[feature_subset[0]] != 0) & (df[feature_subset[1]] != 0)]
    df_x = df[(df[feature_subset[0]] != 0) | (df[feature_subset[1]] != 0)]
    h = plt.hist2d(df_x[feature_subset[0]], df_x[feature_subset[1]], bins=40, norm=LogNorm())
    plt.xlabel(feature_subset[0])
    plt.ylabel(feature_subset[1])
    #plt.colorbar()
    plt.show()
```
